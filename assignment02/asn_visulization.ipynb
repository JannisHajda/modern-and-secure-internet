```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
INPUT_FILE = 'asn_results.csv'
```

```python
df_input = pd.read_csv(INPUT_FILE)
df_input
```

```python
total_websites = df_input['num_http3_websites'].sum()
print(f"Total Number of HTTP/3 Websites: {total_websites}")
```

```python
plt.figure(figsize=(20, 8))

df_input.plot(
    kind='bar',
    x='asn',
    y='num_http3_websites',
    legend=False,
    ax=plt.gca()
)

plt.xlabel('ASN')
plt.ylabel('Number of HTTP/3 Websites')
plt.title('Number of HTTP/3 Websites per ASN')

plt.tight_layout()
plt.show()
```
